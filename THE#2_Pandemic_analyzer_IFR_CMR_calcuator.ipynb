```python

from  tkinter import *
from tkinter.ttk import Combobox
from csv import reader
from tkinter import Entry  
# Creating tkinter window 
window = Tk() 
window.title('PANDEMIC ANALYZER') 
window.geometry('700x500')
imh=PhotoImage(file="corona_a.png")
Label(window,image=imh,bg="snow",bd=0).grid(row=8,column=7)

with open('Data_set.csv', 'r') as read_obj:
    # pass the file object to reader() to get the reader object
    data_set = reader(read_obj)
    # Pass reader object to list() to get a list of lists
    list_of_rows = list(data_set)

# label text for title 
l1=Label(window, text = "PANDEMIC ANALYZER",  background = 'green', foreground ="white",font = ("Times New Roman", 15)).grid(row = 0, column = 1) 
  
# label 
l2=Label(window, text = "Select the State :", font = ("Times New Roman", 10)).grid(column = 0, row = 5, padx = 10, pady = 25) 


l3=Label(window,text="POPULATION:",font = ("Times New Roman", 10)).grid(column = 0, row = 6, padx = 10, pady = 25) 

popu=Entry(window,fg="yellow")


l4=Label(window,text="CONFIRMED:",font = ("Times New Roman", 10)).grid(column = 0, row = 7, padx = 10, pady = 25)

confde=Entry(window,fg="Violet")



l5=Label(window,text="ACTIVE:",font = ("Times New Roman", 10)).grid(column = 0, row = 8, padx = 10, pady = 25)

acte=Entry(window,fg="orange")
acte.grid(column=1,row=8)

l6=Label(window,text="RECOVERED:",font = ("Times New Roman", 10)).grid(column = 0,row = 9, padx = 10, pady = 25)

recd=Entry(window,fg="green")
recd.grid(column=1,row=9)

l7=Label(window,text="DECEASED:",font = ("Times New Roman", 10)).grid(column = 0,row = 10, padx = 10, pady = 25) 

decd=Entry(window,fg="red")
decd.grid(column=1,row=10)

ifr1=Entry(window)
ifr1.grid(row=6,column=7)

cr1=Entry(window)
cr1.grid(row=7,column=7)


  
data1=[]
data2=[]
#Creating List using Loop
for i in range(len(list_of_rows)):
    data1.append(list_of_rows[i])
    data2.append(data1[i][0])

# Creating choose list(combo box)

state_list=Combobox(window,values=data2)
state_list.grid(column = 1, row =5)


#Creating Function for clearing the text

def enable_items():
    
    confde.configure(state='normal')
    acte.configure(state='normal')
    recd.configure(state='normal')
    decd.configure(state='normal')
    popu.configure(state='normal')
    confde.delete(0, 'end')
    acte.delete(0, 'end')
    recd.delete(0, 'end')
    decd.delete(0, 'end')
    popu.delete(0, 'end')  
    cr1.configure(state='normal')
    cr1.delete(0, 'end')
    ifr1.configure(state='normal')
    ifr1.delete(0, 'end')
    
 

def click():
    #print(data1)
    enable_items()

    for i in range(len(data1)):
            #print(data1)
    
            dat=data1[i][0]
            if state_list.get()==dat:
                confde.insert(END,data1[i][1])
                acte.insert(END, data1[i][2])
                recd.insert(END, data1[i][3])
                decd.insert(END, data1[i][4])
                popu.insert(END, data1[i][5])
                confde.configure(state='disabled')
                acte.configure(state='disabled')
                recd.configure(state='disabled')
                decd.configure(state='disabled')
                popu.configure(state='disabled')
def ifr():
        num1=int(confde.get())
        num2=int(acte.get())
        num3=int(recd.get())
        num4=int(decd.get())
        num5=int(popu.get())
        num1==num2+num3+num4
        result=(num4/num1)*100
        ifr1.insert(END, str(result))
        ifr1.configure(state='disabled')
def cmr():
        num1=int(confde.get())
        num2=int(acte.get())
        num3=int(recd.get())
        num4=int(decd.get())
        num5=int(popu.get())
        num1==num2+num3+num4
        result=(num4/num5)*100
        cr1.insert(END, str(result)) 
        cr1.configure(state='disabled')
#section of buttons

cl1=Button(window, text='Click',bg="snow",fg="red",command=click)
cl1.grid(column=5,row=5)

ifr2=Button(window,text='IFR',bg="blue",fg="white",command=ifr)
ifr2.grid(row=6,column=5)

cr2=Button(window,text='CMR',bg="yellow",fg="red",command=cmr)
cr2.grid(row=7,column=5)



        
popu.grid(column=1,row=6)
confde.grid(column=1,row=7)
           



window.mainloop() 
```
