```python
plt.figure(figsize=(8,5))
plt.grid()
plt.plot(maha_framec.Date,maha_framec.MH,'b.-',label="Confirmed")
plt.plot(maha_framed.Date,maha_framed.MH,'r^-',label="Deceased")
plt.plot(maha_framer.Date,maha_framer.MH,'g*-',label="Recovered")
plt.title("Covid19 INSIGHT OF MAHARASHTRA")
plt.xlabel("Date")
plt.ylabel("Confirmed cases")
plt.legend()
plt.xticks(maha_framec.Date[0:30:7].tolist())
plt.show()
```

```python
c=maha_framec.MH.sum()
d=maha_framed.MH.sum()
r=maha_framer.MH.sum()
a=(c)-(d+r)
```

```python
a
```

```python
covid_insight=[c,d,r,a]
inp=['Confirmed','casualties','Recovered','Active']
fig = plt.figure(figsize =(10, 7)) 
plt.pie(covid_insight, labels = inp)
plt.show()
```

```python
# Creating explode data 
explode = (0.1, 0.0, 0.2,0.0) 
  
# Creating color parameters 
colors = ( "orange", "cyan", "brown","blue") 
  
# Wedge properties 
wp = { 'linewidth' : 1, 'edgecolor' : "green" } 
  
# Creating autocpt arguments 
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%\n({:d} g)".format(pct, absolute) 
  
# Creating plot 
fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(covid_insight,  
                                  autopct = lambda pct: func(pct,covid_insight), 
                                  explode = explode,  
                                  labels = inp, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="magenta")) 
  
# Adding legend 
ax.legend(wedges, inp, 
          title ="status", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title("Current Dashboard for bed planning") 
```

```python
WEEK = [0]*len(CovidConf)
for x in range(0,len(CovidConf)):
    WEEK[x] = np.floor(abs(x-len(CovidConf))/7)

CovidConf.insert(0, "WEEK", WEEK, True) 
CovidReco.insert(0, "WEEK", WEEK, True) 
CovidDece.insert(0, "WEEK", WEEK, True) 

CovidConfWEEK = CovidConf.groupby('WEEK').sum().sort_index(axis = 0,ascending=False) 
CovidRecoWEEK = CovidReco.groupby('WEEK').sum().sort_index(axis = 0,ascending=False)  
CovidDeceWEEK = CovidDece.groupby('WEEK').sum().sort_index(axis = 0,ascending=False) 

barWidth = 0.25
r1 = np.arange(len(CovidConfWEEK))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

plt.figure(figsize=(5,5))
plt.bar(r1,CovidConfWEEK['MH'], color = 'b', width = 0.25, label = 'MAHARASHTRA')
plt.bar(r2,CovidConfWEEK['GJ'], color = 'g', width = 0.25, label = 'GUJARAT')
plt.bar(r3,CovidConfWEEK['GA'], color = 'r', width = 0.25, label = 'GOA')
plt.title("WEEKLY Confirmed: Maharashtra/Neighbour States")
plt.xlabel("WEEK")
plt.ylabel("CONFIRMED")
plt.legend()

plt.figure(figsize=(5,5))
plt.bar(r1,CovidRecoWEEK['MH'], color = 'b', width = 0.25, label = 'MAHARASHTRA')
plt.bar(r2,CovidRecoWEEK['GJ'], color = 'g', width = 0.25, label = 'GUJARAT')
plt.bar(r3,CovidRecoWEEK['GA'], color = 'r', width = 0.25, label = 'GOA')
plt.title("WEEKLY Recovered: Maharashtra/Neighbour States")
plt.xlabel("WEEK")
plt.ylabel("RECOVERED")
plt.legend()

plt.figure(figsize=(5,5))
plt.bar(r1,CovidDeceWEEK['MH'], color = 'b', width = 0.25, label = 'MAHARASHTRA')
plt.bar(r2,CovidDeceWEEK['GJ'], color = 'g', width = 0.25, label = 'GUJARAT')
plt.bar(r3,CovidDeceWEEK['GA'], color = 'r', width = 0.25, label = 'GOA')
plt.title("WEEKLY Deceased: Maharashtra/Neighbour States")
plt.xlabel("WEEK")
plt.ylabel("casualties")
plt.legend()
```
